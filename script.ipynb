```python
import pandas as pd
import numpy as np
```

```python
df_pitch = pd.read_excel("Pitchfinder DB.xlsx", usecols = "C, P")
```

```python
df_schools = pd.read_excel("Schools.xlsx", usecols = "E, AU")
```

```python
df_pools = pd.read_excel("Pools.xlsx", usecols = "A,E")
```

```python
df_schools.info()
```

```python
df_pitch['name'] = df_pitch['name'].apply(lambda x: x.lower())
df_pitch['name'] = df_pitch['name'].apply(lambda x: x.replace("'", ""))
df_pitch['name'] = df_pitch['name'].apply(lambda x: x.replace(".", " "))

df_pools['Name'] = df_pools['Name'].apply(lambda x: x.lower())
df_pools['Name'] = df_pools['Name'].apply(lambda x: x.replace("'", ""))
df_pools['Name'] = df_pools['Name'].apply(lambda x: x.replace(".", " "))

df_schools['EstablishmentName'] = df_schools['EstablishmentName'].apply(lambda x: x.lower())
df_schools['EstablishmentName'] = df_schools['EstablishmentName'].apply(lambda x: x.replace("'", ""))
df_schools['EstablishmentName'] = df_schools['EstablishmentName'].apply(lambda x: x.replace(".", " "))
```

```python
df_pitch["name2"] = df_pitch["name"].apply(lambda x: x.split()[0])
df_pitch["name2"] = df_pitch["name2"].apply(lambda x: "st" if x == "saint" else x)

df_pools["Name2"] = df_pools["Name"].apply(lambda x: x.split()[0])
df_pools["Name2"] = df_pools["Name2"].apply(lambda x: "st" if x == "saint" else x)


df_schools["EstablishmentName2"] = df_schools["EstablishmentName"].apply(lambda x: x.split()[0])
df_schools["EstablishmentName2"] = df_schools["EstablishmentName2"].apply(lambda x: "st" if x == "saint" else x)
```

```python
df_master = df_schools.merge(df_pitch, how = "left", left_on = ["Postcode", "EstablishmentName2"], right_on = ["postcode", "name2"])
```

```python
df_final = df_master.merge(df_pools,  how = "left", left_on = ["Postcode", "EstablishmentName2"], right_on = ["Post Code", "Name2"])
```

```python
df_final.head()
```

```python
df_final.drop(columns = ["EstablishmentName2", "postcode", "name2", "Post Code", "Name2"], inplace = True)
```

```python
df_final.rename(columns = {"name":"Pitchfinder_name", "Name":"Pool_name"}, inplace = True)
```
